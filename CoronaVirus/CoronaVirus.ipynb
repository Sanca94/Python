```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
Virus = pd.read_csv('CoronaVirus.csv')
```

```python
Virus = Virus.rename(columns={'2019–20 coronavirus outbreak by country and territory Country or territory [a]': 'Country', 'Recoveries [b]': 'Recoveries'})
```

```python
Virus.head(1000)
```

```python

Virus['Country'] = Virus['Country'].str.replace(" [33]", "")
#Virus['Country'].str.strip('a')
Virus['Country'] =  Virus['Country'].apply(lambda x: x.replace('[','').replace(']',''))
```

```python
Virus['Recoveries'] = Virus['Recoveries'].str.replace("–", "0")
Virus['Recoveries']
```
