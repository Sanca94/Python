# Get Data

```python
df = pd.read_csv('Data.csv')
df.head(46)
```

# Clean Data

```python
df=df.fillna(df.mean()) # any null values that are numbers will be replaced with the mean
df['tl_rank'].mean() 
df.shape
```

# Epcot

```python
epcot = df[df['park_location'] == 'EC'] #MK AK EC HS
epcot
```

```python
epcot['tl_rank']
```

```python
epcot['ride_name']
```

```python
x = ['Frozen Ever After','Gran Fiesta','Journey Into Imagination','Living with the Land','Mission Space','Soarin','Spaceship Earth', 'Test Track']
# epcot['ride_name']
plt.figure(figsize=(15,12))
plt.bar(x,epcot['tl_rank'], color=['black', 'red', 'green', 'blue', 'cyan'])
plt.xlabel('Epcot Attractions')
#plt.xticks(['Spaceship Earth'],['1','2'])
plt.ylabel('Rating')
plt.xticks(fontsize=15, rotation=30,)
plt.title(' Attraction Rating ')
plt.show()
```
